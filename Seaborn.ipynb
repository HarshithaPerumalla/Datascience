```python
import seaborn as sns
```

```python
import pandas as pd
```

Seaborn



```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
# heatmap allows to display corrrelation
correlation- interrelationship is betwen two or more features
```

```python
df.dtypes
```

```python
df_encoded = pd.get_dummies(df, columns=['sex', 'smoker','day','time'])

# Calculate correlations on the encoded DataFrame
df_encoded.corr()
```

```python
sns.heatmap(df_encoded.corr())
```

```python
#joinplot-ubnivariative anakysis

sns.jointplot(x='total_bill', y='tip', data=df,kind='reg')
```

```python
#Pair plot
is also known as scatterplot. in which one variable in the sma edata row is matched with another viariables value. like this.
more than 2 independent features.
so it permutates all the combinations of features
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex') #based on sex feature
```

```python
df['smoker'].value_counts() #tells you to know about counts
```

```python
#distplot
#helps to create the histogram and fit a normal distribution
sns.distplot(df['tip'],kde=False,bins=10) #kde-kernel density estimation
```

```python
#countplot : it picks up the defined coulm and sees how many categories in given cloum and gives the count

sns.countplot(x='smoker',data=df) # vettical obseervation view x='somoker'
```
