Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
print(x,y)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2)
```

```python
#subplots(no.rows,no.column,position) basically tells how canvas is divided for plot
plt.subplot(2,3,1)
plt.plot(x,y,'r--')
plt.subplot(2,3,2)
plt.plot(x,y,'g*-')
plt.subplot(2,3,6)
plt.plot(x,y,'bo')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
#barplots
x=[1,6,9]
y=[89,87,90]

x2=[2,7,4]
y2=[98,45,82]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('x axis')
plt.ylabel('yaxis')
```

```python
# histogram helps you find out the density on y axis
a=np.array([21,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
```

```python
#boxplot used to represent percentile
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
```
